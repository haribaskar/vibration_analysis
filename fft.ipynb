```python
# import numpy and fast fourier transform
import numpy as np
import matplotlib.pyplot as plt
from scipy import pi
from scipy.fftpack import fft
```

```python
# prepare a time domain signal.

# sample_rate is dened as number of samples taken per second.
# Sample rate of 1024 means, 1024 values of the signal are
# recorded in one second

# N is the size of the array. (2 – 0) indicates that only 2 seconds data
# is available. So, if the sample rate is 1024 and we have 2 seconds
# data, the size of the array will be (2 – 0) * 1024 which is 2048.

sample_rate = 1024
N = (2 - 0) *sample_rate
```

```python
time = np.linspace(0, 2, N)
```

```python
# There are two sine waveforms; one with magnitude of 25 at 60Hz
# frequency and other with magnitude of 2 at 270Hz frequency

freq1 = 60
magnitude1 = 25
freq2 = 270
magnitude2 = 2
waveform1 = magnitude1 * np.sin (2 * pi * freq1 * time)
waveform2 = magnitude2 * np.sin (2 * pi * freq2 * time)
```

```python
waveform1.shape,waveform2.shape
```

```python
# add some noise component in our signal.

noise = np.random.normal (0, 3, N)
noise.shape
```

```python
# add the waveforms and noise to make up our time data.

time_data = waveform1 + waveform2 + noise
time_data.shape
```

```python
plt.plot (time [0:100], time_data [0:100])
plt.title ('Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
```

```python
#we have end-point of 512 even though our sampling
# rate was 1024. That’s because according to Nyquist-Shannon
# Sampling theorem, we can only analyze frequency components
# up to half of the sampling rate

frequency = np.linspace (0.0, 512, int (N/2))
freq_data = fft(time_data)
y = 2/N * np.abs (freq_data [0:np.int (N/2)])
```

```python
plt.plot(frequency, y)
plt.title('Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
```
